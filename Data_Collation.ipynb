```python
import pandas as pd
data_list = []
columns = ['route',
 'name',
 'type',
 'departure_time',
 'arrival_time',
 'duration',
 'price',
 'seats_available',
 'rating']

files_list = [
            'Jammu_kashmir_buses.csv',
            'KAAC_buses.csv',
            'West_bengal_buses.csv',
            'NORTH_BENGAL_buses.csv',
            'Bihar_buses.csv',
            'Chandigarh_buses.csv',
            'West_Bengal_ctc_buses.csv',
            'Uttar_Pradesh_buses.csv',
            'Assam_buses.csv',
            'Punjab_buses.csv',
            'south_Bengal_buses.csv',
            'Telangana_buses.csv',
            'Rajasthan_buses.csv',
            'Kerala_buses.csv',
            'Himachal_Pradesh_buses.csv',
            'Goa_buses.csv',
            'Andhra_Pradesh_buses.csv']


for file in files_list:
    df = pd.read_csv(file)
    df = df[columns]
    data_list.append(df)
    
file_data = pd.concat(data_list)
file_data = file_data.reset_index(drop = True)
file_data[['from_place', 'to_place']] = file_data['route'].str.split(' to ', expand=True)
```

```python

data=pd.read_csv("All_Bus_data_Collated.csv")
data
```

```python

def change_name(x):
    return x.str.replace(r'[A-Za-z]+', '', regex=True)
data['price']=data['price'].transform(change_name)
```

```python

data['seats_available']=data['seats_available'].transform(change_name)
data
```

```python
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating']=data['rating'].fillna(0)
data['rating']=data['rating'].astype('float64')
```
